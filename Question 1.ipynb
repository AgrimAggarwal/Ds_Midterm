## Norm
Let $\mathsf{V}$ be a vector space over a field $ \mathbb{F} $ of the real numbers $\mathbb{R}$. A ***norm*** on $\mathsf{V}$ is defined to be a non-negative valued function, $\Vert \cdot \Vert:\mathsf{V}\rightarrow [0,\infty)$ satisfying the following properties :

1. $\forall v\in\mathsf{V}: \Vert v \Vert \geq 0$  and equality holds iff $v= 0 $ (Positive Definite)
2. $\forall v\in\mathsf{V}, a \in \mathbb{R}: \Vert av \Vert = \vert a \vert\Vert v \Vert$ (Absolutely Scalable)
3. $\forall u,v\in\mathsf{V}: \Vert u+v \Vert \leq \Vert u \Vert +\Vert v \Vert$ (Subadditivity or triangle inequality)

For an N-dimensional vector space, ***p-norm*** is definded as follows :
$$ \Vert v \Vert_p = (\sum_{n = 1}^{N}\vert v_n\vert^p)^\frac{1}{p}, \;\;\;\;\;\;\;\;\;\;\;\;\;\; 1\leq p< \infty $$ 

For $p < 1$, this definition of norm is not valid since it violates the triangle equality. We can verify so by looking at the 2-dimensional case for basis vectors, $u = \begin{pmatrix} 1 \\ 0\end{pmatrix} and \: v =\begin{pmatrix} 0 \\ 1\end{pmatrix} $

Calculating LHS of property (3),
$$\begin{align}\Vert u+v \Vert = (\sum_{i = 1}^{2}\vert u_i+v_i \vert^p)^\frac{1}{p}
                  = (\vert 1+0 \vert^p + \vert 0+1 \vert^p)^\frac{1}{p} \end{align}\\\;\;\;\;\;\;\;\;\;\;\;\;
                  = 2^\frac{1}{p}$$

Now calculating RHS of property (3),
$$\Vert u \Vert +\Vert v \Vert = (\sum_{i = 1}^{2}\vert u_i \vert^p)^\frac{1}{p}+(\sum_{i = 1}^{2}\vert v_i \vert^p)^\frac{1}{p}
= 2 $$

Since for $p<1$, $2^\frac{1}{p} > 2$. The triangle inequality does not hold and the norm is not defined.

We can also infer this from the convexity of norms since ***p-norms*** are all convex. To check this we refer to the definition of convex function and prove it for norms also.

A function $f:\mathsf{V}\rightarrow\mathbb{R}$ is said to be convex if $dom(\mathsf{V})$ is a convex set and $\forall u,v \in \mathsf{V}$ and $\forall \lambda \in [0,1]$,

$$ f(\lambda u + (1-\lambda)v) \leq \lambda f(u) + (1-\lambda)f(v)$$

We can see that every p-norm is convex using property (2) and (3) : 

$$ \Vert \lambda u + (1-\lambda)v \Vert \leq \Vert \lambda u \Vert + \Vert (1-\lambda)v \Vert \\
\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;= \vert \lambda \vert \Vert u \Vert + \vert 1-\lambda \vert \Vert v \Vert$$

P-norm for p<1 does not obey this property as can be clearly seen from the unit ball simulation (See below for p=0.5) as the line segment from any two points on the boundary does not lie in the set.

```python
import mt_prng as mt 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

r = mt.mt_seed(5489)
```

```python
'''Function for unit ball simulation. (10,000 x 2) array of random uniform points is generated for which p-norm is less than or
   equal to 1 are plotted to get a unit ball'''

def unitball(p):
    for i in range(10000):
        pt = [mt.random()*2-1, mt.random()*2-1] #Generating 2-d uniform random points in range (-1,1)
        if ((np.abs(pt[0]))**p + (np.abs(pt[1]))**p)**(1/p) <= 1:
            ptt_x.append(pt[0]) #appending x-points for which norm is less than 1
            ptt_y.append(pt[1]) #appending y-points for which norm is less than 1 
        else:
            continue         
```

```python
ptt_x=[]
ptt_y=[]
p =[0.5,1,2,3,7,100,1000] #different p values for p-norm

'''Function to plot unit ball for different p-norms taken from the list above'''
for i in p :
    unitball(i)
    plt.figure()
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.xlim(-1.5,1.5)
    plt.ylim(-1.5,1.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.scatter(ptt_x,ptt_y, label = 'p='+ str(i))
    plt.legend()
    ptt_x=[]
    ptt_y=[]
```
